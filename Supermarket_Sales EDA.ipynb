```python
#who has bought the most products in an city

sales['City'].value_counts()
```

```python
#LETS PLOT THIS IN AN GRAPH
x=list(sales['City'].value_counts().keys())
y=list(sales['City'].value_counts())

plt.bar(x,y)
plt.xlabel('City')
plt.ylabel('Number of sales')
```

```python
sns.histplot(x='Unit _price',bins=20,data=sales,kde=False)
plt.title('Distribution of unit price')
```

```python
#lets see which gender has spent the most in each city
sns.barplot(x='City',y='Total',data=sales,hue='Gender')
```

```python
#find the unique customer type and the frequency of the each type
sales['Customer_ type'].unique()
sales['Customer_ type'].value_counts()
```

```python
#Lets plot which customer_type has bought the most products
sns.countplot('Customer_ type',data=sales)
```

```python
#the total sales of all the products
total_sales=sales['Total'].sum()
total_sales
```

```python
#City with the most sales 
city_with_highest_sales = sales.loc[sales['Total'].idxmax(), 'City']
city_with_highest_sales
```

```python
plt.figure(figsize=(15,15))
sns.barplot(x='Product _line',y='Unit _price',data=sales,hue='City')
```

```python
#the frequency of payments done
sales['Payment'].value_counts()
```

```python
#lets plot this in an pie graph
plt.pie(list(sales['Payment'].value_counts()),labels=list(sales['Payment'].value_counts().keys()),autopct='%0.1f%%')
```
